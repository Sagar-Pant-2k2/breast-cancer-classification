### DATA VALUES

### COST VALUE
#### model => f(x) = M*x+C
##### cost  J(M,C) = summation {(y - M*x+C)^2}

```python
# plt.axis('equal')
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(x,y)
for i in range(1,1000):
    plt.plot(x,list(map(lambda a: a*i,x)))
```
